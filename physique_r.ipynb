# Variation de la tension par rapport au temps

graphique de l'évoluation de la tension et de l'intensité du courant par rapport au temps avec les données suivante :

- Resistance = 20 ohm
- Capacité = 0.5 Nf
- quantité initiale = 70 c
- tension initiale = q0 / c

```python
import datetime
import random
import matplotlib.pyplot as plt
import math

R = 20
C = 0.5 #* math.pow(10, -6)
Q0 = 70
T = R * C

U0 = Q0 / C
I0 = U0 / R

def variationTension(t):
    return (U0 * math.pow(math.e, - t / T))

def variationIntensite(t):
    return (I0 * math.pow(math.e, -t / T))

tensions = []
intensites = []

for i in range(120):
    tensions.append(variationTension(i)) 
    intensites.append(variationIntensite(i))

plt.figure(figsize=(10, 4), dpi=100)
plt.grid(True)
plt.title("Variation de la tension par rapport au temps")
plt.xlabel("t (temps en secondes)")
plt.ylabel("u(t) tension")
plt.bar([*range(0, 120)], tensions)
plt.show()
```

# Variation du courant par rapport au temps

```python
plt.figure(figsize=(10, 4), dpi=100)
plt.grid(True)
plt.title("Variation de l'intensité par rapport au temps")
plt.xlabel("t (temps en secondes)")
plt.ylabel("i(t) intensité")
plt.bar([*range(0, 120)], intensites)
plt.show()
```
