# Variation du courant par rapport temps (circuit RL)

graphique de la variation du courant par rapport au temps dans un circuit RL avec les donnÃ©es suivantes :

- L = 0.8H
- R0 = 8 Ohm
- R = 1K Ohm
- I0 = E/Ro => 12 / 8 => 1.5 A

```python
import matplotlib.pyplot as plt
import math

L = 0.8
I0 = 1.5
R0 = 0.008
R = 1

def variationIntensite(t):
    Re = R + R0
    return I0 * math.pow(math.e, - (Re/L) * t)

intensites = []
for i in range(120):
    intensites.append(variationIntensite(i))

            
plt.figure(figsize=(10, 4), dpi=100)
plt.grid(True)
plt.title("Variation de l'intensitÃ© par rapport au temps (circuit RL)")
plt.xlabel("t (temps en secondes)")
plt.ylabel("i(t) intensitÃ©")
plt.plot([*range(0, 120)], intensites)
plt.show()
```
