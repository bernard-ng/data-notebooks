# Régime d'un Circuit RLC

visualisation avec les données suivante :
- r1 = 20 Ω, r2 = 15 Ω, r3 = 10 Ω, r4 = 4 Ω, r5 = 1 Ω
- wo = 10 rad / sec
- A = 2, B = 1, L = 1


```python
import matplotlib.pyplot as plt
import math

# Données de depart
wo = 10; A = 2; B = 1; L =1
r1 = 20; r2 = 15; r3 = 10; r4 = 4; r5 = 1

# Définition des fonctions (formules)
def Q(r):
    return (L * wo) / r

def W(Q):
    return wo * math.sqrt(((1 / (4 * math.pow(Q, 2))) - 1) * -1)

def T(Q):
    return (2 * Q) / wo

def i(t, T, W):
    return math.pow(math.e, (-t / T)) * (A * math.cos(W * t) + B * math.sin(W * t))

# Affection des valeurs calculées
Q1 = Q(r1); Q2 = Q(r2); Q3 = Q(r3); Q4 = Q(r4); Q5 = Q(r5)
w1 = W(Q1); w2 = W(Q2); w3 = W(Q3); w4 = W(Q4); w5 = W(Q5)
T1 = T(Q1); T2 = T(Q2); T3 = T(Q3); T4 = T(Q4); T5 = T(Q5)
i1 = []; i2 = []; i3 = []; i4 = []; i5 = []

# Calcule pour 15 secondes
temps = [*range(0, 10)]
for t in range(10):
    i1.append(i(t, T1, w1))
    i2.append(i(t, T2, w2))
    i3.append(i(t, T3, w3))
    i4.append(i(t, T4, w4))
    i5.append(i(t, T5, w5))


plt.figure(figsize=(8,4), dpi=100, tight_layout=True)
plt.grid(True)
plt.title("Variation de l'intensité par rapport au temps (circuit RLC)")
plt.xlabel("t (temps en secondes)")
plt.ylabel("i(t) intensité")

plt.plot(temps, i1, color="#84cc16", label="courant i1")
plt.plot(temps, i2, color="#18181b", label="courant i2")
plt.plot(temps, i3, color="#d946ef", label="courant i3")
plt.plot(temps, i4, color="#22d3ee", label="courant i4")
plt.plot(temps, i5, color="#f97316", label="courant i5")

plt.legend(title='Intensités')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8,4), dpi=100, tight_layout=True)

gs = fig.add_gridspec(2, 2, hspace=0, wspace=0)
(ax1, ax2), (ax3, ax4) = gs.subplots(sharex='col', sharey='row')

fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
plt.grid(False)
plt.xlabel("t (temps en secondes)")
plt.ylabel("i(t) intensité")

fig.suptitle('Amortissement pseudo-périodique pour chaque courant (i)')

ax1.plot(temps, i2, color="#84cc16", label="courant i1, i2")
ax2.plot(temps, i3, color="#d946ef", label="courant i3")
ax3.plot(temps, i4, color="#22d3ee", label="courant i4")
ax4.plot(temps, i5, color="#f97316", label="courant i5")

ax2.legend(); ax1.legend(); ax3.legend(); ax4.legend()

for ax in axs.flat:
    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)
    ax.margins(-0.4, 0.0)
    ax.label_outer()
```
